```python
# We need to check a list of numbers, where each number should be between 0 and n (with n being the length of the list).
#  We need to find any numbers in that range that are missing from the list. If no numbers are missing, we return -1.
```


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
# [2,0,4,5,4] return [1,3]


# We have the list [1, 5, 10].

# The list length is 3.
# The smallest number is 1, and the largest is 10.
# Since the list length isn’t 1 or 2, we proceed with the regular process.
# We start checking from the number after the smallest one, which is 2, and continue until we reach the largest number (10), but we stop before 10 itself.

# Here’s what happens step by step:

# 2: Not in the list, so we add it to the missing numbers.
# 3: Missing as well, so add it.
# 4: Missing, add it.
# 5: Found in the list, so skip it.
# 6: Missing, add it.
# 7: Missing, add it.
# 8: Missing, add it.
# 9: Missing, add it.
# We stop when we reach 10, as we only check numbers smaller than the largest one.

# The missing numbers we found are: [2, 3, 4, 6, 7, 8, 9].
```


-   Copy the solution your partner wrote. 


```python
def missing_num(nums: list) -> list[int]:
  result = []
    
  len_lst = len(nums)
  max_lst = max(nums)
  min_lst = min(nums)
  
  #base case

  if len_lst == 1:
    return -1
  elif len_lst == 2 and max_lst == min_lst + 1:
    return -1
  else:
    
    i = min_lst + 1
    while i < max_lst:
      if i not in nums:
        result.append(i)
      i = i + 1
    if result:
      return result
    else:
      return -1
```


-   Explain why their solution works in your own words.


```python
# The code works like this:

# Figure out the range: It checks for missing numbers between the smallest and largest numbers in the list, starting from the smallest number plus one
#  and stopping just before the largest number.

# Check every number in the range: For each number in that range, it checks if the number is in the list. If it’s not there, the number gets added to the list of missing numbers.

# Handle special cases: If the list has only one number or has two numbers that are right next to each other, the code assumes there are no missing numbers and returns -1.

# Return the result: If no missing numbers are found after checking everything, the code returns -1. Otherwise, it returns the list of missing numbers.

# In short, the code goes through all the numbers that could be in the list and collects the ones that aren’t. If no numbers are missing, it returns -1.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time Complexity:
# The code checks each number between the smallest and largest values in the list to see if it’s missing. 
# In Python, checking if a number is in a list takes time that grows with the size of the list because it looks through the list one item at a time. 
# If there’s a big gap between the smallest and largest numbers, the code has to repeat this check for each number in that gap. 
# This means that, in the worst case, the time it takes can grow quite a lot; up to n² when the list and the gap between numbers are both large.

# Space Complexity:
# The code keeps a list of all the missing numbers it finds. In the worst-case scenario, where a lot of numbers are missing, 
# this list could end up being almost as large as the original one. So, the amount of extra space the code needs grows in proportion to the size of the list, 
# making the space complexity O(n).
```


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


```python
# The solution checks for missing numbers between the smallest and largest numbers in the list. 
# However, in many versions of this problem, you might need to consider the entire range from 0 up to n (where n is the length of the list). 
# If the list has numbers greater than n, the largest value could exceed n, which would change the range being checked. 
# The current solution focuses only on the gap between the smallest and largest numbers in the list. This approach works well for finding missing numbers in 
# that specific range but might not fully match the problem if it specifically requires finding missing numbers from [0..n].

# Time Complexity:
# The code checks if each number is in the list, which takes time proportional to the list size each time it runs. In the worst case, this process can take O(n²) time. 
# A more efficient approach would be to store all the list’s numbers in a set before checking. Checking membership in a set takes constant time (O(1)),
#  which would bring the overall time complexity down to O(n) when iterating through the range.

# Edge Cases:

# If the list is empty, using min() or max() would cause an error. Usually, the problem assumes at least one element, but it’s something to keep in mind.
# If the list contains repeated numbers, the solution still works correctly but will run more slowly because of repeated membership checks. 
# Using a set would make these checks more efficient.
# In summary:
# The solution’s logic is correct, but it could be made faster by using a set for membership checks. 
# Also, clarifying whether the full [0..n] range or just the min-to-max range should be considered would help ensure the solution matches the problem’s requirements.
```
