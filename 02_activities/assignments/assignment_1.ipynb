
-   Code the solution to your assigned problem in Python (code chunk). Try to find the best time and space complexity solution!


```python
from collections import deque

def is_duplicate(root):
    if not root:
        return -1
    
    seen = set()
    queue = deque([root])
    
    while queue:
        node = queue.popleft()
        
        if node.val in seen:
            return node.val
        seen.add(node.val)
        
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    
    return -1
```


-   Explain why your solution works


```python
# We use a queue to perform a breadth-first search, which starts from the root.
# We keep a set called seen to keep track of the values we've already seen.
# As soon as we see a value that is already in the set, we know we found a duplicate. Because BFS processes nodes level by level from the top down,
#  the first duplicate we see is definitely the closest one to the root.
# If the search finishes without finding any duplicates, we return -1.
```


-   Explain the problem’s time and space complexity


```python
# Time Complexity: O(n) because we have to check every node in the tree.
# Space Complexity: O(n) because we use a queue to store the nodes level by level.
```


-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


```python
# Initializing the first level: We start by placing only the root node into a list called current_level.
# Maintaining a seen set: We keep a set named seen to record values we've already seen.
# Processing each level: For every node in current_level, we check whether its value is already in seen.
# If it is, we return that value immediately (we've found our first duplicate).
# If not, we add the value to seen and collect its children for processing in the following level.
# Moving to the next level: Once we're done checking all nodes in current_level, we set current_level to next_level—the children we gathered and then repeat.
# If we finish processing all levels without finding duplicates, we return -1.
```
